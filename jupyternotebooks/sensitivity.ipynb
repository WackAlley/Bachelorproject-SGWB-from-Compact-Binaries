```python
# Sensitivity curves
# Gabriella Agazie et al 2023 ApJL 951 L10 (DOI 10.3847/2041-8213/acda88)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_table(
    'sensitivity_curves/sensitivity_curves_NG15yr_fullPTA.txt',
    sep=','
)
df
```

```python
# renaming
df.columns = ['frequency','strain','signal','density']
df = df.set_index('frequency')
df
```

```python
fig, ax = plt.subplots()
ax.plot(df.index, df['density'],linewidth=.5, label=f'NANOGrav detector curve, $\\Omega$')
ax.plot(df.index, df['strain'],linewidth=.5, label=f'NANOGrav detector curve, $h_{{c}}$')
ax.plot(df.index, df['signal'],linewidth=.5, label=f'NANOGrav detector curve, S')
ax.set_yscale('log')
ax.set_xscale('log')
ax.legend(loc='upper left', fontsize=8)
plt.show()
```
