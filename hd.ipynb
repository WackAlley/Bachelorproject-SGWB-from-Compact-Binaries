```python
import numpy as np
import matplotlib.pyplot as plt

def hellings_downs_correlation(theta):
    """
    Compute the Hellings-Downs correlation for a given angular separation.
    
    Parameters:
    - theta (float or ndarray): Angular separation in radians (0 to pi).
    
    Returns:
    - hd_corr (float or ndarray): Hellings-Downs correlation values.
    """
    # Avoid numerical issues near theta = 0 or pi
    small_angle = 1e-10
    theta = np.clip(theta, small_angle, np.pi - small_angle)

    # Hellings-Downs correlation formula
    hd_corr = (1.0 / 2) * (1 + np.cos(theta)) * np.log(1 + np.cos(theta)) \
              - (1.0 / 12) * (1 + np.cos(theta)) \
              + 1.0/3

    return hd_corr

theta_degrees = np.linspace(0, 180, 500)
theta_radians = np.radians(theta_degrees)
hd_values = hellings_downs_correlation(theta_radians)

plt.figure(figsize=(8, 6))
plt.plot(theta_degrees, hd_values, label="Hellings-Downs Correlation", color="blue")
plt.xticks(np.arange(0, 181, 20))
plt.xlabel("Angular Separation (degrees)")
plt.ylabel("Correlation")
plt.title("Hellings-Downs Correlation Curve")
plt.legend()
plt.grid()
plt.show()
```
